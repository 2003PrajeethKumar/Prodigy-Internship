```python
import matplotlib.pyplot as plt
import numpy as np

# Generate example age data (random)
np.random.seed(1)
age_data = np.random.randint(18, 80, size=1000)

# Create histogram
plt.hist(age_data, bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
